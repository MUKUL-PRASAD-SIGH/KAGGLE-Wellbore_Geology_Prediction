```python
# ╔══════════════════════════════════════════════════════════════╗
# ║   ROGII – Wellbore Geology Prediction  |  v4 memory-safe    ║
# ║   Three rules:                                               ║
# ║   1. float32 everywhere  (halves RAM vs float64)            ║
# ║   2. one well at a time  (never load all features at once)  ║
# ║   3. lean feature set    (no redundant lag explosions)      ║
# ╚══════════════════════════════════════════════════════════════╝

import os, gc, glob, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
import lightgbm as lgb

import os
import glob
import warnings

warnings.filterwarnings("ignore")

# ──────────────────────────────────────────────
# PATHS
# ──────────────────────────────────────────────

BASE = "/kaggle/input/competitions/rogii-wellbore-geology-prediction"

TRAIN_H    = f"{BASE}/train/*__horizontal_well.csv"
TRAIN_T    = f"{BASE}/train/*__typewell.csv"

TEST_H     = f"{BASE}/test/*__horizontal_well.csv"
TEST_T     = f"{BASE}/test/*__typewell.csv"

SAMPLE_SUB = f"{BASE}/sample_submission.csv"

# ──────────────────────────────────────────────
# DEBUG
# ──────────────────────────────────────────────

print("BASE EXISTS:", os.path.exists(BASE))

print("\nCompetition folder:")
print(os.listdir("/kaggle/input/competitions"))

print("\nTrain files:")
print(glob.glob(TRAIN_H)[:5])

print("\nTrain count:")
print(len(glob.glob(TRAIN_H)))
# ──────────────────────────────────────────────
# RULE 1 — downcast helper
# Every column cast to float32 / int16 right away.
# Calling this once saves ~50 % RAM instantly.
# ──────────────────────────────────────────────
def downcast(df):
    for c in df.select_dtypes("float").columns:
        df[c] = df[c].astype(np.float32)
    for c in df.select_dtypes("integer").columns:
        mx = df[c].abs().max()
        df[c] = df[c].astype(np.int16 if mx < 32_767 else np.int32)
    return df

def mb(df):
    return df.memory_usage(deep=True).sum() / 1024**2


# ──────────────────────────────────────────────
# LOAD  (downcast on the fly, inside the loop)
# ──────────────────────────────────────────────
def load_wells(pattern, split):
    files = sorted(glob.glob(pattern))
    out   = []
    for f in files:
        df            = pd.read_csv(f, low_memory=False)
        df["WELL_NAME"] = os.path.basename(f).split("__")[0]
        df["SPLIT"]     = split
        df.columns      = [c.strip().upper() for c in df.columns]
        df = downcast(df)          # ← RULE 1 applied immediately
        out.append(df)
    combined = pd.concat(out, ignore_index=True)
    print(f"  {split:5s}  {len(files):>4} files  "
          f"{len(combined):>9,} rows  {mb(combined):>6.0f} MB")
    return combined

print("Loading …")
h_train = load_wells(TRAIN_H, "train")
t_train = load_wells(TRAIN_T, "train")
h_test  = load_wells(TEST_H,  "test")
t_test  = load_wells(TEST_T,  "test")

# Keep only useful columns (drop nothing critical, drop junk)
KEEP = ["WELL_NAME", "MD", "X", "Y", "Z",
        "GR", "TVT", "TVT_INPUT",
        "ANCC", "ASTNU", "ASTNL", "EGFDU", "EGFDL", "BUDA"]

h_train = h_train[[c for c in KEEP if c in h_train.columns]]
h_test  = h_test [[c for c in KEEP if c in h_test.columns]]
gc.collect()

print(f"\nAfter trim →  train {mb(h_train):.0f} MB   test {mb(h_test):.0f} MB")


# ──────────────────────────────────────────────
# BUILD TYPEWELL LOOKUP  (tiny dict, not a df)
# ──────────────────────────────────────────────
# ──────────────────────────────────────────────
# BUILD TYPEWELL LOOKUP
# ──────────────────────────────────────────────

print("\nTypewell columns:")
print(t_train.columns.tolist())

# determine depth column automatically
TW_DEPTH = "MD" if "MD" in t_train.columns else "TVT"

print("Using typewell depth column:", TW_DEPTH)

TW_SIGS = [
    c for c in t_train.columns
    if c not in {"WELL_NAME", "SPLIT", TW_DEPTH}
    and t_train[c].dtype != object
]

tw_lookup = {}

for wname, g in t_train.groupby("WELL_NAME"):

    g = g.sort_values(TW_DEPTH).reset_index(drop=True)

    tw_lookup[wname] = (
        g[[TW_DEPTH] + TW_SIGS]
        .values
        .astype(np.float32)
    )

# global fallback
tw_global = (
    t_train
    .sort_values(TW_DEPTH)
    .reset_index(drop=True)[[TW_DEPTH] + TW_SIGS]
    .values
    .astype(np.float32)
)

del t_train, t_test
gc.collect()


# ──────────────────────────────────────────────
# RULE 2 + 3 — feature engineering ONE WELL
# Process one well at a time → peak RAM stays
# tiny.  Feature set is intentionally lean.
# ──────────────────────────────────────────────
SIG_COLS = [c for c in
            ["GR","ANCC","ASTNU","ASTNL","EGFDU","EGFDL","BUDA","TVT_INPUT"]
            if c in h_train.columns]

LAGS   = [1, 3, 5]     # only 3 lag values
ROLL_W = 5             # only 1 rolling window size

def featurize(df, well_name):
    """
    df   : rows for ONE well, sorted by MD
    Returns the same rows + feature columns, all float32.
    """
    n   = len(df)
    out = {}   # build as dict → concat once at end (faster)

    # copy original columns
    for c in df.columns:
        out[c] = df[c].values.copy()

    for col in SIG_COLS:
        if col not in df.columns:
            continue
        arr = df[col].values.astype(np.float32)

        # lags & leads
        for k in LAGS:
            lag          = np.full(n, np.nan, dtype=np.float32)
            lead         = np.full(n, np.nan, dtype=np.float32)
            lag[k:]      = arr[:-k]
            lead[:-k]    = arr[k:]
            out[f"{col}_L{k}"] = lag
            out[f"{col}_F{k}"] = lead

        # rolling mean + std  (one window)
        s = pd.Series(arr)
        out[f"{col}_RM5"] = s.rolling(ROLL_W, min_periods=1).mean().values.astype(np.float32)
        out[f"{col}_RS5"] = s.rolling(ROLL_W, min_periods=1).std() .values.astype(np.float32)

        # first difference
        d    = np.full(n, np.nan, dtype=np.float32)
        d[1:] = arr[1:] - arr[:-1]
        out[f"{col}_D1"] = d

    # MD step
    md = df["MD"].values.astype(np.float32)
    ms    = np.full(n, np.nan, dtype=np.float32)
    ms[1:] = md[1:] - md[:-1]
    out["MD_STEP"] = ms

    # 3-D step distance + cumulative
    if all(c in df.columns for c in ["X","Y","Z"]):
        dx = np.concatenate([[np.nan], np.diff(df["X"].values.astype(np.float32))])
        dy = np.concatenate([[np.nan], np.diff(df["Y"].values.astype(np.float32))])
        dz = np.concatenate([[np.nan], np.diff(df["Z"].values.astype(np.float32))])
        step = np.sqrt(dx**2 + dy**2 + dz**2).astype(np.float32)
        out["STEP3D"]   = step
        out["CUM_DIST"] = np.nancumsum(step).astype(np.float32)

    # TVT_INPUT extras
    if "TVT_INPUT" in df.columns:
        ti = pd.Series(df["TVT_INPUT"].values.astype(np.float32))
        out["TI_FF"]     = ti.ffill().values.astype(np.float32)
        out["TI_BF"]     = ti.bfill().values.astype(np.float32)
        out["TI_ISNAN"]  = ti.isna().astype(np.int8).values
        # how many consecutive NaN rows since last known value
        run = np.zeros(n, dtype=np.int16)
        cnt = np.int16(0)
        for i in range(n):
            cnt = np.int16(cnt + 1) if np.isnan(ti.iloc[i]) else np.int16(0)
            run[i] = cnt
        out["TI_RUN"] = run

    # position inside well  0 → 1
    mn_md, mx_md = float(md.min()), float(md.max())
    out["WELL_PROG"] = ((md - mn_md) / (mx_md - mn_md + 1e-9)).astype(np.float32)

    # typewell nearest-MD lookup
    tw = tw_lookup.get(well_name, tw_global)
    idx = np.searchsorted(tw[:, 0], md).clip(0, len(tw) - 1)
    for j, sig in enumerate(TW_SIGS, start=1):
        out[f"TW_{sig}"] = tw[idx, j].astype(np.float32)

    return pd.DataFrame(out)


def engineer_all(h_df):
    """
    RULE 2: loop over wells one at a time.
    Peak RAM = one well in memory at a time.
    """
    wells  = h_df["WELL_NAME"].unique()
    pieces = []
    for i, wname in enumerate(wells, 1):
        mask = h_df["WELL_NAME"] == wname
        wdf  = h_df[mask].sort_values("MD").reset_index(drop=True)
        pieces.append(featurize(wdf, wname))
        if i % 150 == 0:
            print(f"  {i}/{len(wells)} wells done …")
    result = pd.concat(pieces, ignore_index=True)
    gc.collect()
    return result


print("\nEngineering train features …")
fe_train = engineer_all(h_train)
print(f"Train FE done  {fe_train.shape}  {mb(fe_train):.0f} MB")

print("\nEngineering test features …")
fe_test  = engineer_all(h_test)
print(f"Test  FE done  {fe_test.shape}   {mb(fe_test):.0f} MB")

del h_train, h_test
gc.collect()


# ──────────────────────────────────────────────
# TRAIN / PREDICT SPLIT
# ──────────────────────────────────────────────
TARGET = "TVT"

known   = fe_train[fe_train[TARGET].notna()].copy()
df_pred = fe_test.copy()

print(f"\nTraining rows : {len(known):,}")
print(f"Test rows     : {len(df_pred):,}")

# Encode well name
le = LabelEncoder()
all_wells = pd.concat([known["WELL_NAME"], df_pred["WELL_NAME"]]).astype(str)
le.fit(all_wells)
known  ["WELL_ID"] = le.transform(known  ["WELL_NAME"].astype(str)).astype(np.int32)
df_pred["WELL_ID"] = le.transform(df_pred["WELL_NAME"].astype(str)).astype(np.int32)

# Build feature list
EXCLUDE = {TARGET, "TVT_INPUT", "SPLIT", "WELL_NAME"}
FEATURE_COLS = [c for c in known.columns
                if c not in EXCLUDE
                and known[c].dtype != object]
FEATURE_COLS = list(dict.fromkeys(FEATURE_COLS))

print(f"Feature count : {len(FEATURE_COLS)}")

X      = known  [FEATURE_COLS].astype(np.float32)
y      = known  [TARGET].astype(np.float32)
X_pred = df_pred[[c for c in FEATURE_COLS if c in df_pred.columns]].copy()

# add any missing columns as zero
for c in FEATURE_COLS:
    if c not in X_pred.columns:
        X_pred[c] = np.float32(0)
X_pred = X_pred[FEATURE_COLS].astype(np.float32)

print(f"X {X.shape}  {mb(X):.0f} MB")
print(f"X_pred {X_pred.shape}  {mb(X_pred):.0f} MB")
gc.collect()


# ──────────────────────────────────────────────
# LIGHTGBM  5-FOLD CV
# ──────────────────────────────────────────────
def rmse(a, b):
    return float(np.sqrt(mean_squared_error(
        np.asarray(a), np.asarray(b))))

N   = 5
kf  = KFold(n_splits=N, shuffle=True, random_state=42)

params = dict(
    objective         = "regression",
    metric            = "rmse",
    device            = "gpu",
    num_leaves        = 63,
    learning_rate     = 0.05,
    feature_fraction  = 0.80,
    bagging_fraction  = 0.80,
    bagging_freq      = 5,
    min_child_samples = 50,
    n_estimators      = 3000,
    reg_alpha         = 0.05,
    reg_lambda        = 0.1,
    verbose           = -1,
    n_jobs            = -1,
    random_state      = 42,
)

oof   = np.zeros(len(X),      dtype=np.float32)
preds = np.zeros(len(X_pred), dtype=np.float32)
fold_scores = []

print("\nLightGBM CV …")
for fold, (tr, vl) in enumerate(kf.split(X, y), 1):
    m = lgb.LGBMRegressor(**params)
    m.fit(
        X.iloc[tr], y.iloc[tr],
        eval_set=[(X.iloc[vl], y.iloc[vl])],
        callbacks=[lgb.early_stopping(150, verbose=False),
                   lgb.log_evaluation(500)],
    )
    vp          = m.predict(X.iloc[vl]).astype(np.float32)
    oof[vl]     = vp
    preds      += m.predict(X_pred).astype(np.float32) / N
    r           = rmse(y.iloc[vl], vp)
    fold_scores.append(r)
    print(f"  Fold {fold}  RMSE={r:.5f}  iter={m.best_iteration_}")
    gc.collect()

oof_score = rmse(y, oof)
print(f"\nCV  : {np.mean(fold_scores):.5f} ± {np.std(fold_scores):.5f}")
print(f"OOF : {oof_score:.5f}")


# ──────────────────────────────────────────────
# PLOTS
# ──────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

fi = (pd.DataFrame({"f": FEATURE_COLS, "i": m.feature_importances_})
      .sort_values("i", ascending=False).head(25))
axes[0].barh(fi["f"][::-1], fi["i"][::-1], color="#1565C0")
axes[0].set_title("Top-25 Feature Importances")

step = max(1, len(y) // 50_000)   # plot at most 50k points
axes[1].scatter(y.values[::step], oof[::step],
                alpha=0.2, s=3, color="#B71C1C")
mn, mx = float(y.min()), float(y.max())
axes[1].plot([mn, mx],[mn, mx],"k--",lw=1)
axes[1].set_title(f"OOF Pred vs Actual  RMSE={oof_score:.4f}")
axes[1].set_xlabel("Actual TVT"); axes[1].set_ylabel("Predicted TVT")

res = y.values - oof
axes[2].hist(res, bins=80, color="#1B5E20", edgecolor="white", lw=0.3)
axes[2].axvline(0, color="black", lw=1.5)
axes[2].set_title("OOF Residuals")

plt.tight_layout()
plt.savefig("diagnostics.png", dpi=110)
plt.show()


# ──────────────────────────────────────────────
# SUBMISSION
# ──────────────────────────────────────────────
print("\nBuilding submission.csv …")
sample = pd.read_csv(SAMPLE_SUB)
print("Columns in sample_submission:", sample.columns.tolist())
print(sample.head(3))

id_col  = sample.columns[0]
tvt_col = sample.columns[1]

# Strategy A – explicit ID column in test set
id_cands = [c for c in fe_test.columns
            if c.lower() in {"id","row_id","uid","index"}]
if id_cands:
    sub = pd.DataFrame({id_col: df_pred[id_cands[0]].values,
                        tvt_col: preds})

# Strategy B – sample has WELL_NAME + MD
elif {"WELL_NAME","MD"}.issubset(sample.columns):
    pk = df_pred[["WELL_NAME","MD"]].copy()
    pk[tvt_col] = preds
    sub = sample[[id_col,"WELL_NAME","MD"]].merge(
        pk, on=["WELL_NAME","MD"], how="left"
    )[[id_col, tvt_col]]

# Strategy C – row order
else:
    print("  ⚠  row-order match — check sample_submission format")
    sub = pd.DataFrame({id_col:  sample[id_col].values,
                        tvt_col: preds[:len(sample)]})

assert len(sub) == len(sample), \
    f"Row mismatch: {len(sub)} vs {len(sample)}"
assert sub[tvt_col].isna().sum() == 0, \
    f"{sub[tvt_col].isna().sum()} NaN predictions"

sub.to_csv("submission.csv", index=False)
print("\n✅  submission.csv ready!")
print(sub.head(10))
print(f"\nStats: mean={preds.mean():.4f}  std={preds.std():.4f}  "
      f"min={preds.min():.4f}  max={preds.max():.4f}")
print(f"\n  OOF RMSE = {oof_score:.5f}")
print("  Upload submission.csv to the leaderboard 🚀")
```

```python
# ╔══════════════════════════════════════════════════════════════════╗
# ║   ROGII – Wellbore Geology Prediction  |  v5 FINAL             ║
# ║                                                                  ║
# ║   FIXES vs v4:                                                   ║
# ║   ★ ROW_IDX preserved → submission alignment FIXED (907→~5)    ║
# ║   ★ GPU LightGBM (device='gpu', max_bin=63)                    ║
# ║   ★ Early stopping per fold (no over-training)                  ║
# ║   ★ Multi-seed (3 seeds × 5 folds) ensemble                    ║
# ║   ★ Richer features (EWM, D2, multi-window rolling, TW diffs)  ║
# ║   ★ Parquet caching → skip FE on re-runs                       ║
# ║   ★ Hard submission assertions before writing                   ║
# ╚══════════════════════════════════════════════════════════════════╝

import os, gc, glob, warnings, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
import lightgbm as lgb

warnings.filterwarnings("ignore")

# ─────────────────────────────────────────────────────────────────
# CONFIG
# ─────────────────────────────────────────────────────────────────

BASE       = "/kaggle/input/competitions/rogii-wellbore-geology-prediction"
WORK       = "/kaggle/working"

TRAIN_H    = f"{BASE}/train/*__horizontal_well.csv"
TRAIN_T    = f"{BASE}/train/*__typewell.csv"
TEST_H     = f"{BASE}/test/*__horizontal_well.csv"
TEST_T     = f"{BASE}/test/*__typewell.csv"
SAMPLE_SUB = f"{BASE}/sample_submission.csv"

FE_TRAIN_CACHE = f"{WORK}/fe_train_v5.parquet"
FE_TEST_CACHE  = f"{WORK}/fe_test_v5.parquet"

USE_CACHE  = True   # set False to force FE recompute

SEEDS      = [42, 123, 777]   # 3 seeds → averaged predictions
N_FOLD     = 5


# ─────────────────────────────────────────────────────────────────
# HELPERS
# ─────────────────────────────────────────────────────────────────

def downcast(df: pd.DataFrame) -> pd.DataFrame:
    """Cast floats → float32, ints → int16/32.  Halves RAM instantly."""
    for c in df.select_dtypes("float").columns:
        df[c] = df[c].astype(np.float32)
    for c in df.select_dtypes("integer").columns:
        mx = df[c].abs().max()
        df[c] = df[c].astype(np.int16 if mx < 32_767 else np.int32)
    return df

def mb(df: pd.DataFrame) -> float:
    return df.memory_usage(deep=True).sum() / 1024 ** 2

def rmse(a, b) -> float:
    return float(np.sqrt(mean_squared_error(np.asarray(a), np.asarray(b))))


# ─────────────────────────────────────────────────────────────────
# STEP 1 ─ LOAD WELLS
# ★ ROW_IDX ADDED BEFORE CONCAT — THIS IS THE CRITICAL BUG FIX ★
#
#   Without ROW_IDX, Strategy-C (row-order) submission mapping was
#   used → predictions attached to wrong geology rows → LB = 907.
#   With ROW_IDX we construct IDs as "WELLNAME_rowindex" which
#   matches the sample_submission ID format exactly.
# ─────────────────────────────────────────────────────────────────

def load_wells(pattern: str, split: str) -> pd.DataFrame:
    files = sorted(glob.glob(pattern))
    out   = []
    for f in files:
        df = pd.read_csv(f, low_memory=False)

        # ★ CRITICAL: capture row position INSIDE this file before concat
        df["ROW_IDX"]   = np.arange(len(df), dtype=np.int32)
        df["WELL_NAME"] = os.path.basename(f).split("__")[0]
        df["SPLIT"]     = split

        # normalise column names
        df.columns = [c.strip().upper() for c in df.columns]

        df = downcast(df)
        out.append(df)

    combined = pd.concat(out, ignore_index=True)
    print(f"  {split:5s}  {len(files):>4d} files  "
          f"{len(combined):>9,} rows  {mb(combined):>6.0f} MB")
    return combined


# ─────────────────────────────────────────────────────────────────
# STEP 2 ─ TYPEWELL LOOKUP
# ─────────────────────────────────────────────────────────────────

def build_tw_lookup(t_df: pd.DataFrame):
    """Build a per-well dict of sorted typewell arrays for fast lookup."""
    TW_DEPTH = "MD" if "MD" in t_df.columns else "TVT"
    TW_SIGS  = [
        c for c in t_df.columns
        if c not in {"WELL_NAME", "SPLIT", TW_DEPTH}
        and t_df[c].dtype != object
    ]
    tw_lookup: dict = {}
    for wname, g in t_df.groupby("WELL_NAME"):
        g = g.sort_values(TW_DEPTH).reset_index(drop=True)
        tw_lookup[wname] = g[[TW_DEPTH] + TW_SIGS].values.astype(np.float32)

    # global fallback (concatenated + sorted)
    tw_global = (
        t_df.sort_values(TW_DEPTH).reset_index(drop=True)
        [[TW_DEPTH] + TW_SIGS].values.astype(np.float32)
    )
    return tw_lookup, tw_global, TW_SIGS, TW_DEPTH


# ─────────────────────────────────────────────────────────────────
# STEP 3 ─ FEATURE ENGINEERING (one well at a time → low peak RAM)
# ─────────────────────────────────────────────────────────────────

LAGS   = [1, 2, 3, 5, 10]
ROLL_W = [3, 5, 10]


def featurize(df: pd.DataFrame, well_name: str,
              tw_lookup, tw_global, TW_SIGS, TW_DEPTH,
              SIG_COLS) -> pd.DataFrame:
    """
    Build all features for a single well.
    Keeps df small → peak RAM = one well at a time.
    """
    n   = len(df)
    out = {}

    # ── copy every original column (includes ROW_IDX, WELL_NAME, MD …) ──
    for c in df.columns:
        out[c] = df[c].values.copy()

    # ── signal-level features ──
    for col in SIG_COLS:
        if col not in df.columns:
            continue
        arr = df[col].values.astype(np.float32)
        s   = pd.Series(arr)

        # lags and leads
        for k in LAGS:
            lag        = np.full(n, np.nan, dtype=np.float32)
            lead       = np.full(n, np.nan, dtype=np.float32)
            lag[k:]    = arr[:-k]
            lead[:-k]  = arr[k:]
            out[f"{col}_L{k}"] = lag
            out[f"{col}_F{k}"] = lead

        # rolling stats — multiple windows
        for w in ROLL_W:
            rm = s.rolling(w, min_periods=1)
            out[f"{col}_RM{w}"]  = rm.mean().values.astype(np.float32)
            out[f"{col}_RS{w}"]  = rm.std().values.astype(np.float32)
            out[f"{col}_RMIN{w}"]= rm.min().values.astype(np.float32)
            out[f"{col}_RMAX{w}"]= rm.max().values.astype(np.float32)

        # exponential weighted mean
        out[f"{col}_EWM5"] = s.ewm(span=5, min_periods=1).mean().values.astype(np.float32)

        # first derivative (depth-space)
        d1     = np.full(n, np.nan, dtype=np.float32)
        d1[1:] = arr[1:] - arr[:-1]
        out[f"{col}_D1"] = d1

        # second derivative
        d2     = np.full(n, np.nan, dtype=np.float32)
        d2[2:] = d1[2:] - d1[1:-1]
        out[f"{col}_D2"] = d2

        # deviation from 5-point rolling mean (local anomaly)
        rm5 = out.get(f"{col}_RM5",
                      s.rolling(5, min_periods=1).mean().values.astype(np.float32))
        out[f"{col}_DEV5"] = (arr - rm5).astype(np.float32)

    # ── MD-based features ──
    md     = df["MD"].values.astype(np.float32)
    ms     = np.full(n, np.nan, dtype=np.float32)
    ms[1:] = md[1:] - md[:-1]
    out["MD_STEP"]      = ms
    out["MD_FROM_START"]= (md - md[0]).astype(np.float32)

    mn_md, mx_md = float(md.min()), float(md.max())
    out["WELL_PROG"] = ((md - mn_md) / (mx_md - mn_md + 1e-9)).astype(np.float32)
    out["WELL_LEN"]  = np.full(n, mx_md - mn_md, dtype=np.float32)

    # ── 3-D geometry ──
    if all(c in df.columns for c in ["X", "Y", "Z"]):
        x_v = df["X"].values.astype(np.float32)
        y_v = df["Y"].values.astype(np.float32)
        z_v = df["Z"].values.astype(np.float32)

        dx   = np.concatenate([[np.nan], np.diff(x_v)])
        dy   = np.concatenate([[np.nan], np.diff(y_v)])
        dz   = np.concatenate([[np.nan], np.diff(z_v)])
        step = np.sqrt(dx**2 + dy**2 + dz**2).astype(np.float32)
        out["STEP3D"]   = step
        out["CUM_DIST"] = np.nancumsum(step).astype(np.float32)

        # inclination angle (degrees from vertical)
        horiz = np.sqrt(dx**2 + dy**2)
        with np.errstate(invalid="ignore", divide="ignore"):
            angle = np.degrees(np.arctan2(horiz, np.abs(dz))).astype(np.float32)
        angle[~np.isfinite(angle)] = np.nan
        out["INCL_ANGLE"] = angle

    # ── TVT_INPUT imputation features ──
    if "TVT_INPUT" in df.columns:
        ti = pd.Series(df["TVT_INPUT"].values.astype(np.float32))

        ff = ti.ffill().values.astype(np.float32)
        bf = ti.bfill().values.astype(np.float32)
        it = ti.interpolate(method="linear",
                            limit_direction="both").values.astype(np.float32)

        out["TI_FF"]     = ff
        out["TI_BF"]     = bf
        out["TI_INTERP"] = it
        out["TI_ISNAN"]  = ti.isna().astype(np.int8).values

        # gradient of forward-filled TVT_INPUT
        ti_d1      = np.full(n, np.nan, dtype=np.float32)
        ti_d1[1:]  = ff[1:] - ff[:-1]
        out["TI_GRAD"] = ti_d1

        # consecutive NaN run length
        run = np.zeros(n, dtype=np.int16)
        cnt = np.int16(0)
        for i in range(n):
            cnt   = np.int16(cnt + 1) if np.isnan(ti.iloc[i]) else np.int16(0)
            run[i] = cnt
        out["TI_RUN"] = run

    # ── typewell nearest-MD lookup ──
    tw  = tw_lookup.get(well_name, tw_global)
    idx = np.searchsorted(tw[:, 0], md).clip(0, len(tw) - 1)

    for j, sig in enumerate(TW_SIGS, start=1):
        tw_vals = tw[idx, j].astype(np.float32)
        out[f"TW_{sig}"] = tw_vals

        # difference: measured − typewell (captures local deviation)
        if sig in df.columns:
            out[f"TW_{sig}_DIFF"] = (
                df[sig].values.astype(np.float32) - tw_vals
            ).astype(np.float32)

    return pd.DataFrame(out)


def engineer_all(h_df: pd.DataFrame,
                 tw_lookup, tw_global, TW_SIGS, TW_DEPTH,
                 SIG_COLS) -> pd.DataFrame:
    """Loop over wells one at a time to keep peak RAM minimal."""
    wells  = h_df["WELL_NAME"].unique()
    pieces = []
    for i, wname in enumerate(wells, 1):
        mask = h_df["WELL_NAME"] == wname
        wdf  = h_df[mask].sort_values("MD").reset_index(drop=True)
        pieces.append(featurize(wdf, wname,
                                tw_lookup, tw_global,
                                TW_SIGS, TW_DEPTH, SIG_COLS))
        if i % 100 == 0:
            print(f"    {i}/{len(wells)} wells done …")
    result = pd.concat(pieces, ignore_index=True)
    gc.collect()
    return result


# ═════════════════════════════════════════════════════════════════
#  MAIN PIPELINE
# ═════════════════════════════════════════════════════════════════

print("=" * 66)
print("  ROGII Wellbore Geology Prediction  |  v5 FINAL")
print("=" * 66)

# ── 1. LOAD ───────────────────────────────────────────────────────
print("\n[1] Loading well CSV files …")
h_train = load_wells(TRAIN_H, "train")
t_train = load_wells(TRAIN_T, "train")
h_test  = load_wells(TEST_H,  "test")
t_test  = load_wells(TEST_T,  "test")

# ── 2. TYPEWELL LOOKUP ────────────────────────────────────────────
print("\n[2] Building typewell lookup …")
tw_lookup, tw_global, TW_SIGS, TW_DEPTH = build_tw_lookup(t_train)
print(f"    Typewell signals: {TW_SIGS}")
del t_train, t_test
gc.collect()

# ── 3. TRIM COLUMNS ───────────────────────────────────────────────
# ★ ROW_IDX is NOW included in KEEP — it was missing before ★
KEEP = [
    "WELL_NAME", "ROW_IDX",               # identity / submission key
    "MD", "X", "Y", "Z",                  # geometry
    "GR", "TVT", "TVT_INPUT",             # logging / target
    "ANCC", "ASTNU", "ASTNL",
    "EGFDU", "EGFDL", "BUDA",
]
h_train = h_train[[c for c in KEEP if c in h_train.columns]]
h_test  = h_test [[c for c in KEEP if c in h_test.columns]]

SIG_COLS = [c for c in
            ["GR", "ANCC", "ASTNU", "ASTNL",
             "EGFDU", "EGFDL", "BUDA", "TVT_INPUT"]
            if c in h_train.columns]
print(f"\n[3] Signal columns: {SIG_COLS}")
print(f"    Train after trim: {h_train.shape}   "
      f"Test: {h_test.shape}")

# ── 4. FEATURE ENGINEERING (with parquet caching) ─────────────────
cache_valid = (USE_CACHE
               and os.path.exists(FE_TRAIN_CACHE)
               and os.path.exists(FE_TEST_CACHE))

if cache_valid:
    # quick check: caches must contain ROW_IDX
    _cols = pd.read_parquet(FE_TEST_CACHE, columns=["ROW_IDX"]).columns
    if "ROW_IDX" not in _cols:
        print("\n[4] Cache missing ROW_IDX — recomputing …")
        cache_valid = False

if cache_valid:
    print("\n[4] Loading FE from parquet cache …")
    fe_train = pd.read_parquet(FE_TRAIN_CACHE)
    fe_test  = pd.read_parquet(FE_TEST_CACHE)
    print(f"    train {fe_train.shape}   test {fe_test.shape}")
else:
    print("\n[4] Engineering train features …")
    fe_train = engineer_all(h_train, tw_lookup, tw_global,
                            TW_SIGS, TW_DEPTH, SIG_COLS)
    print(f"    Train FE: {fe_train.shape}  {mb(fe_train):.0f} MB")

    print("\n    Engineering test features …")
    fe_test = engineer_all(h_test, tw_lookup, tw_global,
                           TW_SIGS, TW_DEPTH, SIG_COLS)
    print(f"    Test  FE: {fe_test.shape}  {mb(fe_test):.0f} MB")

    print("    Saving to parquet cache …")
    fe_train.to_parquet(FE_TRAIN_CACHE, index=False)
    fe_test .to_parquet(FE_TEST_CACHE,  index=False)
    print("    Cached!")

del h_train, h_test
gc.collect()

# ── 5. PREPARE ARRAYS ─────────────────────────────────────────────
TARGET  = "TVT"
# ★ ROW_IDX in EXCLUDE → present in df but never used as predictor ★
EXCLUDE = {TARGET, "TVT_INPUT", "SPLIT", "WELL_NAME", "ROW_IDX"}

known   = fe_train[fe_train[TARGET].notna()].copy()
df_pred = fe_test.copy()

print(f"\n[5] Training rows: {len(known):,}   Test rows: {len(df_pred):,}")

# encode well name as an integer feature
le        = LabelEncoder()
all_names = pd.concat([known["WELL_NAME"],
                       df_pred["WELL_NAME"]]).astype(str)
le.fit(all_names)
known  ["WELL_ID"] = le.transform(known  ["WELL_NAME"].astype(str)).astype(np.int32)
df_pred["WELL_ID"] = le.transform(df_pred["WELL_NAME"].astype(str)).astype(np.int32)

FEATURE_COLS = [c for c in known.columns
                if c not in EXCLUDE
                and known[c].dtype != object]
FEATURE_COLS = list(dict.fromkeys(FEATURE_COLS))   # deduplicate, preserve order

print(f"    Feature count: {len(FEATURE_COLS)}")

X      = known  [FEATURE_COLS].astype(np.float32)
y      = known  [TARGET].astype(np.float32)

X_pred = df_pred[[c for c in FEATURE_COLS if c in df_pred.columns]].copy()
for c in FEATURE_COLS:                              # fill any missing cols
    if c not in X_pred.columns:
        X_pred[c] = np.float32(0)
X_pred = X_pred[FEATURE_COLS].astype(np.float32)

print(f"    X {X.shape}  ({mb(X):.0f} MB)    "
      f"X_pred {X_pred.shape}  ({mb(X_pred):.0f} MB)")
gc.collect()

# ── 6. LIGHTGBM  3-SEED × 5-FOLD CV ──────────────────────────────
#
#  GPU PARAMS
#  • device='gpu'       → GPU kernel
#  • max_bin=63         → LightGBM docs recommend small max_bin for GPU
#  • num_leaves=127     → richer trees than v4's 63
#  • learning_rate=0.03 → slower but deeper convergence; early-stop compensates
#  • n_estimators=5000  → ceiling; early-stopping picks the true optimum
#  • early_stopping=200 → halt if no improvement for 200 rounds
#
#  SEED AVERAGING reduces variance without extra compute cost.
# ─────────────────────────────────────────────────────────────────

BASE_PARAMS = dict(
    objective         = "regression",
    metric            = "rmse",
    device            = "gpu",
    gpu_platform_id   = 0,
    gpu_device_id     = 0,
    max_bin           = 63,
    num_leaves        = 127,
    learning_rate     = 0.03,
    feature_fraction  = 0.75,
    bagging_fraction  = 0.75,
    bagging_freq      = 5,
    min_child_samples = 30,
    n_estimators      = 5000,
    reg_alpha         = 0.05,
    reg_lambda        = 0.10,
    verbose           = -1,
    n_jobs            = -1,
)

oof_accum  = np.zeros(len(X),      dtype=np.float64)
pred_accum = np.zeros(len(X_pred), dtype=np.float64)
total_runs = 0
all_fold_scores = []

print(f"\n[6] LightGBM  {N_FOLD}-fold × {len(SEEDS)} seeds …\n")

for seed in SEEDS:
    params   = {**BASE_PARAMS, "random_state": seed, "seed": seed}
    kf       = KFold(n_splits=N_FOLD, shuffle=True, random_state=seed)
    oof_seed = np.zeros(len(X),      dtype=np.float32)
    prd_seed = np.zeros(len(X_pred), dtype=np.float32)

    for fold, (tr_idx, vl_idx) in enumerate(kf.split(X, y), 1):
        t0 = time.time()
        m  = lgb.LGBMRegressor(**params)
        m.fit(
            X.iloc[tr_idx], y.iloc[tr_idx],
            eval_set=[(X.iloc[vl_idx], y.iloc[vl_idx])],
            callbacks=[
                lgb.early_stopping(200, verbose=False),
                lgb.log_evaluation(1000),
            ],
        )
        vp               = m.predict(X.iloc[vl_idx]).astype(np.float32)
        oof_seed[vl_idx] = vp
        prd_seed        += m.predict(X_pred).astype(np.float32) / N_FOLD
        r                = rmse(y.iloc[vl_idx], vp)
        all_fold_scores.append(r)
        elapsed          = time.time() - t0
        print(f"  seed={seed}  fold={fold}  "
              f"RMSE={r:.5f}  iter={m.best_iteration_}  t={elapsed:.0f}s")
        gc.collect()

    seed_oof = rmse(y, oof_seed)
    print(f"  ── seed {seed}  OOF = {seed_oof:.5f}\n")

    oof_accum  += oof_seed.astype(np.float64)
    pred_accum += prd_seed.astype(np.float64)
    total_runs += 1

# averaged predictions
oof_accum  /= total_runs
pred_accum /= total_runs
oof_final   = oof_accum.astype(np.float32)
preds       = pred_accum.astype(np.float32)

oof_score = rmse(y, oof_final)
print(f"\n{'='*66}")
print(f"  CV  : {np.mean(all_fold_scores):.5f} ± {np.std(all_fold_scores):.5f}  "
      f"(mean ± std over all folds/seeds)")
print(f"  OOF : {oof_score:.5f}  (multi-seed averaged)")
print(f"{'='*66}")

# ── 7. DIAGNOSTIC PLOTS ───────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(17, 4))

fi = (pd.DataFrame({"f": FEATURE_COLS, "i": m.feature_importances_})
      .sort_values("i", ascending=False).head(25))
axes[0].barh(fi["f"][::-1], fi["i"][::-1], color="#1565C0")
axes[0].set_title("Top-25 Feature Importances")
axes[0].tick_params(axis="y", labelsize=7)

step = max(1, len(y) // 50_000)
axes[1].scatter(y.values[::step], oof_final[::step],
                alpha=0.2, s=2, color="#B71C1C")
mn, mx = float(y.min()), float(y.max())
axes[1].plot([mn, mx], [mn, mx], "k--", lw=1)
axes[1].set_title(f"OOF Pred vs Actual  RMSE={oof_score:.4f}")
axes[1].set_xlabel("Actual TVT")
axes[1].set_ylabel("Predicted TVT")

res = y.values - oof_final
axes[2].hist(res, bins=80, color="#1B5E20", edgecolor="white", lw=0.3)
axes[2].axvline(0, color="black", lw=1.5)
axes[2].set_title("OOF Residuals")

plt.tight_layout()
plt.savefig("diagnostics.png", dpi=110)
plt.show()
print("  Saved diagnostics.png")


# ─────────────────────────────────────────────────────────────────
# STEP 8 ─ SUBMISSION  (FIXED ID ALIGNMENT)
#
#  Before (v4):   row-order match → predictions on wrong rows → LB 907
#  Now     (v5):  exact WELLNAME_rowindex match → correct mapping
#
#  ID format confirmed from sample_submission:  WELLNAME_rowindex
#  e.g.  000d7d20_1442   →  WELL_NAME="000d7d20"  ROW_IDX=1442
# ─────────────────────────────────────────────────────────────────

print("\n[8] Building submission.csv …")

sample  = pd.read_csv(SAMPLE_SUB)
id_col  = sample.columns[0]
tvt_col = sample.columns[1]
print(f"    Sample columns : {sample.columns.tolist()}")
print(f"    Sample rows    : {len(sample):,}")
print(sample.head(3))

# ★ construct WELLNAME_rowindex IDs exactly
df_pred = df_pred.copy()
df_pred[id_col] = (
    df_pred["WELL_NAME"].astype(str)
    + "_"
    + df_pred["ROW_IDX"].astype(str)
)

sub_raw = pd.DataFrame({
    id_col:  df_pred[id_col].values,
    tvt_col: preds,
})

# align EXACTLY to sample order via merge
sub = sample[[id_col]].merge(sub_raw, on=id_col, how="left")

n_missing = int(sub[tvt_col].isna().sum())
if n_missing > 0:
    fill_val = float(np.nanmean(preds))
    sub[tvt_col].fillna(fill_val, inplace=True)
    print(f"    ⚠  Filled {n_missing} missing predictions "
          f"with global mean = {fill_val:.4f}")
else:
    print(f"    ✓  All {len(sub):,} predictions matched — zero missing")

# ── hard assertions before writing ──────────────────────────────
assert len(sub) == len(sample), \
    f"Row count mismatch: {len(sub)} vs {len(sample)}"
assert sub[tvt_col].isna().sum() == 0, \
    "NaN predictions remain after fill"
assert list(sub.columns) == [id_col, tvt_col], \
    f"Wrong columns: {sub.columns.tolist()}"

sub.to_csv("submission.csv", index=False)

print("\n✅  submission.csv ready!")
print(sub.head(10).to_string())
print(f"\n  Pred stats ─ mean={preds.mean():.4f}  "
      f"std={preds.std():.4f}  "
      f"min={preds.min():.4f}  "
      f"max={preds.max():.4f}")
print(f"\n  Final OOF RMSE = {oof_score:.5f}")

print("""
╔══════════════════════════════════════════════════════════════╗
║  PRE-SUBMISSION CHECKLIST                                    ║
║  ✓  submission.csv exists with correct id + tvt columns     ║
║  ✓  IDs = WELLNAME_ROW_IDX  (matches sample_submission)     ║
║  ✓  Row count matches sample_submission                      ║
║  ✓  Zero NaN predictions                                     ║
║  ⚠  Turn Internet OFF in Kaggle settings before Submit      ║
║  ⚠  Ensure total runtime < 9 hrs  (GPU helps a lot)         ║
╚══════════════════════════════════════════════════════════════╝
""")
print("  🚀  Upload submission.csv to the leaderboard!")
```

```python
# ╔══════════════════════════════════════════════════════════════════╗
# ║  ROGII – Wellbore Geology Prediction  |  v6  lean+correct       ║
# ║                                                                  ║
# ║  WHY v5 OOM-crashed:  ~250 feature cols × 3-5 M rows = 3-5 GB  ║
# ║  v4 was ~108 cols and ran fine.  v6 targets ≤ 110 cols.         ║
# ║                                                                  ║
# ║  WHAT v6 FIXES vs v4:                                           ║
# ║  ★ ROW_IDX preserved  →  submission IDs correct  (907 → ~5)    ║
# ║  ★ TW_sig_DIFF  (measured − typewell)  — one extra per signal   ║
# ║  ★ TI_INTERP  replaces clunky ffill/bfill pair                  ║
# ║  ★ Parquet cache  →  skip FE on re-runs                         ║
# ║  ★ Hard assertions before writing submission                     ║
# ║  ★ Single seed (3 is fine later, but 1 seed won't OOM first)    ║
# ╚══════════════════════════════════════════════════════════════════╝

import os, gc, glob, warnings, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
import lightgbm as lgb

warnings.filterwarnings("ignore")

# ─────────────────────────────────────────────────────────────────
# CONFIG  ── tweak here, nowhere else
# ─────────────────────────────────────────────────────────────────

BASE       = "/kaggle/input/competitions/rogii-wellbore-geology-prediction"
WORK       = "/kaggle/working"

TRAIN_H    = f"{BASE}/train/*__horizontal_well.csv"
TRAIN_T    = f"{BASE}/train/*__typewell.csv"
TEST_H     = f"{BASE}/test/*__horizontal_well.csv"
TEST_T     = f"{BASE}/test/*__typewell.csv"
SAMPLE_SUB = f"{BASE}/sample_submission.csv"

FE_TRAIN_CACHE = f"{WORK}/fe_train_v6.parquet"
FE_TEST_CACHE  = f"{WORK}/fe_test_v6.parquet"
USE_CACHE      = True    # False = recompute FE from scratch

N_FOLD   = 5
SEED     = 42

# ── lean feature knobs  (these keep RAM under control) ────────────
LAGS   = [1, 3, 5]      # 3 lag values  ← same as v4
ROLL_W = 5              # single window ← same as v4


# ─────────────────────────────────────────────────────────────────
# HELPERS
# ─────────────────────────────────────────────────────────────────

def downcast(df: pd.DataFrame) -> pd.DataFrame:
    for c in df.select_dtypes("float").columns:
        df[c] = df[c].astype(np.float32)
    for c in df.select_dtypes("integer").columns:
        mx = df[c].abs().max()
        df[c] = df[c].astype(np.int16 if mx < 32_767 else np.int32)
    return df

def mb(df: pd.DataFrame) -> float:
    return df.memory_usage(deep=True).sum() / 1024 ** 2

def rmse(a, b) -> float:
    return float(np.sqrt(mean_squared_error(
        np.asarray(a), np.asarray(b))))


# ─────────────────────────────────────────────────────────────────
# LOAD
# ★  ROW_IDX must be added BEFORE concat — this is the OOM-free
#    replacement for the v4 row-order mismatch that caused LB=907
# ─────────────────────────────────────────────────────────────────

def load_wells(pattern: str, split: str) -> pd.DataFrame:
    files = sorted(glob.glob(pattern))
    out   = []
    for f in files:
        df = pd.read_csv(f, low_memory=False)

        # ★ CRITICAL: row position inside THIS file, before concat
        df["ROW_IDX"]   = np.arange(len(df), dtype=np.int32)
        df["WELL_NAME"] = os.path.basename(f).split("__")[0]
        df["SPLIT"]     = split
        df.columns      = [c.strip().upper() for c in df.columns]
        df = downcast(df)
        out.append(df)

    combined = pd.concat(out, ignore_index=True)
    print(f"  {split:5s}  {len(files):>4d} files  "
          f"{len(combined):>9,} rows  {mb(combined):>6.0f} MB")
    return combined


# ─────────────────────────────────────────────────────────────────
# TYPEWELL LOOKUP
# ─────────────────────────────────────────────────────────────────

def build_tw_lookup(t_df: pd.DataFrame):
    TW_DEPTH = "MD" if "MD" in t_df.columns else "TVT"
    TW_SIGS  = [
        c for c in t_df.columns
        if c not in {"WELL_NAME", "SPLIT", TW_DEPTH}
        and t_df[c].dtype != object
    ]
    tw_lookup: dict = {}
    for wname, g in t_df.groupby("WELL_NAME"):
        g = g.sort_values(TW_DEPTH).reset_index(drop=True)
        tw_lookup[wname] = (
            g[[TW_DEPTH] + TW_SIGS].values.astype(np.float32)
        )
    tw_global = (
        t_df.sort_values(TW_DEPTH).reset_index(drop=True)
        [[TW_DEPTH] + TW_SIGS].values.astype(np.float32)
    )
    return tw_lookup, tw_global, TW_SIGS, TW_DEPTH


# ─────────────────────────────────────────────────────────────────
# FEATURE ENGINEERING  — lean, one well at a time
#
# Column budget (8 signals):
#   lags+leads   : 8 × 3 × 2  = 48
#   roll mean+std: 8 × 2       = 16
#   D1           : 8           =  8
#   TW lookup    : ~10 sigs    = 10
#   TW diff      : ~10 sigs    = 10  ← new vs v4 (cheap, useful)
#   MD/XYZ misc  :             = 10
#   TVT_INPUT    :             =  6
#   WELL_PROG etc:             =  2
#   WELL_ID etc  :             =  3
#   ────────────────────────────────
#   TOTAL                      ≈ 113   (vs v5's ~250 — safe for 16 GB)
# ─────────────────────────────────────────────────────────────────

def featurize(df: pd.DataFrame, well_name: str,
              tw_lookup, tw_global, TW_SIGS, SIG_COLS) -> pd.DataFrame:
    n   = len(df)
    out = {}

    # copy every original column (ROW_IDX, WELL_NAME, MD, TVT …)
    for c in df.columns:
        out[c] = df[c].values.copy()

    # ── signal features ──────────────────────────────────────────
    for col in SIG_COLS:
        if col not in df.columns:
            continue
        arr = df[col].values.astype(np.float32)
        s   = pd.Series(arr)

        # lags & leads  (3 each → 6 cols per signal)
        for k in LAGS:
            lag       = np.full(n, np.nan, dtype=np.float32)
            lead      = np.full(n, np.nan, dtype=np.float32)
            lag[k:]   = arr[:-k]
            lead[:-k] = arr[k:]
            out[f"{col}_L{k}"] = lag
            out[f"{col}_F{k}"] = lead

        # rolling mean + std  (1 window)
        rm = s.rolling(ROLL_W, min_periods=1)
        out[f"{col}_RM{ROLL_W}"] = rm.mean().values.astype(np.float32)
        out[f"{col}_RS{ROLL_W}"] = rm.std() .values.astype(np.float32)

        # first difference
        d1     = np.full(n, np.nan, dtype=np.float32)
        d1[1:] = arr[1:] - arr[:-1]
        out[f"{col}_D1"] = d1

    # ── MD features ──────────────────────────────────────────────
    md     = df["MD"].values.astype(np.float32)
    ms     = np.full(n, np.nan, dtype=np.float32)
    ms[1:] = md[1:] - md[:-1]
    out["MD_STEP"] = ms

    mn_md, mx_md = float(md.min()), float(md.max())
    out["WELL_PROG"] = ((md - mn_md) / (mx_md - mn_md + 1e-9)).astype(np.float32)

    # ── 3-D step distance ─────────────────────────────────────────
    if all(c in df.columns for c in ["X", "Y", "Z"]):
        dx   = np.concatenate([[np.nan], np.diff(df["X"].values.astype(np.float32))])
        dy   = np.concatenate([[np.nan], np.diff(df["Y"].values.astype(np.float32))])
        dz   = np.concatenate([[np.nan], np.diff(df["Z"].values.astype(np.float32))])
        step = np.sqrt(dx**2 + dy**2 + dz**2).astype(np.float32)
        out["STEP3D"]   = step
        out["CUM_DIST"] = np.nancumsum(step).astype(np.float32)

    # ── TVT_INPUT imputation ──────────────────────────────────────
    if "TVT_INPUT" in df.columns:
        ti = pd.Series(df["TVT_INPUT"].values.astype(np.float32))

        interp = ti.interpolate(method="linear",
                                limit_direction="both").values.astype(np.float32)
        out["TI_INTERP"] = interp
        out["TI_ISNAN"]  = ti.isna().astype(np.int8).values

        # consecutive-NaN run length  (cheap, useful)
        run = np.zeros(n, dtype=np.int16)
        cnt = np.int16(0)
        for i in range(n):
            cnt    = np.int16(cnt + 1) if np.isnan(ti.iloc[i]) else np.int16(0)
            run[i] = cnt
        out["TI_RUN"] = run

    # ── typewell lookup + diff ────────────────────────────────────
    tw  = tw_lookup.get(well_name, tw_global)
    idx = np.searchsorted(tw[:, 0], md).clip(0, len(tw) - 1)

    for j, sig in enumerate(TW_SIGS, start=1):
        tw_vals = tw[idx, j].astype(np.float32)
        out[f"TW_{sig}"] = tw_vals

        # ★ new: how far does the measured signal deviate from typewell?
        if sig in df.columns:
            out[f"TW_{sig}_DIFF"] = (
                df[sig].values.astype(np.float32) - tw_vals
            ).astype(np.float32)

    return pd.DataFrame(out)


def engineer_all(h_df: pd.DataFrame,
                 tw_lookup, tw_global, TW_SIGS, TW_DEPTH,
                 SIG_COLS) -> pd.DataFrame:
    """One well at a time — peak RAM = one well, not all wells."""
    wells  = h_df["WELL_NAME"].unique()
    pieces = []
    for i, wname in enumerate(wells, 1):
        mask = h_df["WELL_NAME"] == wname
        wdf  = h_df[mask].sort_values("MD").reset_index(drop=True)
        pieces.append(featurize(wdf, wname,
                                tw_lookup, tw_global,
                                TW_SIGS, SIG_COLS))
        if i % 150 == 0:
            print(f"    {i}/{len(wells)} wells …")
    result = pd.concat(pieces, ignore_index=True)
    gc.collect()
    return result


# ═════════════════════════════════════════════════════════════════
#  MAIN  PIPELINE
# ═════════════════════════════════════════════════════════════════

print("=" * 66)
print("  ROGII v6  —  lean + correct  (target OOF < 5)")
print("=" * 66)

# ── 1. LOAD ───────────────────────────────────────────────────────
print("\n[1] Loading …")
h_train = load_wells(TRAIN_H, "train")
t_train = load_wells(TRAIN_T, "train")
h_test  = load_wells(TEST_H,  "test")
t_test  = load_wells(TEST_T,  "test")

# ── 2. TYPEWELL LOOKUP ────────────────────────────────────────────
print("\n[2] Typewell lookup …")
tw_lookup, tw_global, TW_SIGS, TW_DEPTH = build_tw_lookup(t_train)
print(f"    signals: {TW_SIGS}")
del t_train, t_test
gc.collect()

# ── 3. TRIM ───────────────────────────────────────────────────────
# ★ ROW_IDX is in KEEP so it survives the column trim
KEEP = [
    "WELL_NAME", "ROW_IDX",               # identity (never used as feature)
    "MD", "X", "Y", "Z",
    "GR", "TVT", "TVT_INPUT",
    "ANCC", "ASTNU", "ASTNL",
    "EGFDU", "EGFDL", "BUDA",
]
h_train = h_train[[c for c in KEEP if c in h_train.columns]]
h_test  = h_test [[c for c in KEEP if c in h_test.columns]]
gc.collect()

SIG_COLS = [c for c in
            ["GR", "ANCC", "ASTNU", "ASTNL", "EGFDU", "EGFDL", "BUDA", "TVT_INPUT"]
            if c in h_train.columns]
print(f"\n[3] Signals: {SIG_COLS}")
print(f"    After trim  train={h_train.shape}  test={h_test.shape}")

# ── 4. FEATURE ENGINEERING (with parquet cache) ───────────────────
cache_ok = (USE_CACHE
            and os.path.exists(FE_TRAIN_CACHE)
            and os.path.exists(FE_TEST_CACHE))

if cache_ok:
    # verify cache has ROW_IDX
    _tmp = pd.read_parquet(FE_TEST_CACHE, columns=["ROW_IDX"])
    if "ROW_IDX" not in _tmp.columns:
        print("\n[4] Cache missing ROW_IDX — rebuilding …"); cache_ok = False
    del _tmp

if cache_ok:
    print("\n[4] Loading FE from cache …")
    fe_train = pd.read_parquet(FE_TRAIN_CACHE)
    fe_test  = pd.read_parquet(FE_TEST_CACHE)
    print(f"    train {fe_train.shape}  test {fe_test.shape}")
else:
    print("\n[4] Engineering train features …")
    fe_train = engineer_all(h_train, tw_lookup, tw_global,
                            TW_SIGS, TW_DEPTH, SIG_COLS)
    print(f"    train FE: {fe_train.shape}  {mb(fe_train):.0f} MB")

    print("\n    Engineering test features …")
    fe_test = engineer_all(h_test, tw_lookup, tw_global,
                           TW_SIGS, TW_DEPTH, SIG_COLS)
    print(f"    test  FE: {fe_test.shape}  {mb(fe_test):.0f} MB")

    fe_train.to_parquet(FE_TRAIN_CACHE, index=False)
    fe_test .to_parquet(FE_TEST_CACHE,  index=False)
    print("    Cached to parquet.")

del h_train, h_test
gc.collect()

# ── 5. PREPARE X / y ──────────────────────────────────────────────
TARGET  = "TVT"
# ★ ROW_IDX in EXCLUDE → stays in df_pred but never reaches model
EXCLUDE = {TARGET, "TVT_INPUT", "SPLIT", "WELL_NAME", "ROW_IDX"}

known   = fe_train[fe_train[TARGET].notna()].copy()
df_pred = fe_test.copy()

print(f"\n[5] Train rows: {len(known):,}   Test rows: {len(df_pred):,}")

le = LabelEncoder()
le.fit(pd.concat([known["WELL_NAME"], df_pred["WELL_NAME"]]).astype(str))
known  ["WELL_ID"] = le.transform(known  ["WELL_NAME"].astype(str)).astype(np.int32)
df_pred["WELL_ID"] = le.transform(df_pred["WELL_NAME"].astype(str)).astype(np.int32)

FEATURE_COLS = [c for c in known.columns
                if c not in EXCLUDE
                and known[c].dtype != object]
FEATURE_COLS = list(dict.fromkeys(FEATURE_COLS))
print(f"    Feature count: {len(FEATURE_COLS)}")

X      = known  [FEATURE_COLS].astype(np.float32)
y      = known  [TARGET].astype(np.float32)

X_pred = df_pred[[c for c in FEATURE_COLS if c in df_pred.columns]].copy()
for c in FEATURE_COLS:
    if c not in X_pred.columns:
        X_pred[c] = np.float32(0)
X_pred = X_pred[FEATURE_COLS].astype(np.float32)

print(f"    X {X.shape}  {mb(X):.0f} MB")
print(f"    X_pred {X_pred.shape}  {mb(X_pred):.0f} MB")
gc.collect()

# ── 6. LIGHTGBM  5-FOLD CV ────────────────────────────────────────
#
#  GPU strategy:
#   device='gpu', max_bin=63  →  fast histogram on GPU
#   num_leaves=127            →  richer than v4 (was 63), still safe
#   learning_rate=0.04        →  balanced; early-stop finds optimum
#   n_estimators=4000         →  ceiling; early_stopping(200) cuts it
#   single SEED               →  saves RAM vs 3-seed loop
#
#  To add more seeds later:
#     run this block 2 more times with seeds 123 and 777,
#     average the preds arrays before building submission.

params = dict(
    objective         = "regression",
    metric            = "rmse",
    device            = "gpu",
    gpu_platform_id   = 0,
    gpu_device_id     = 0,
    max_bin           = 63,
    num_leaves        = 127,
    learning_rate     = 0.04,
    feature_fraction  = 0.80,
    bagging_fraction  = 0.80,
    bagging_freq      = 5,
    min_child_samples = 30,
    n_estimators      = 4000,
    reg_alpha         = 0.05,
    reg_lambda        = 0.10,
    verbose           = -1,
    n_jobs            = -1,
    random_state      = SEED,
    seed              = SEED,
)

kf          = KFold(n_splits=N_FOLD, shuffle=True, random_state=SEED)
oof         = np.zeros(len(X),      dtype=np.float32)
preds       = np.zeros(len(X_pred), dtype=np.float32)
fold_scores = []

print(f"\n[6] LightGBM {N_FOLD}-fold CV  (single seed={SEED}) …\n")
for fold, (tr_idx, vl_idx) in enumerate(kf.split(X, y), 1):
    t0 = time.time()
    m  = lgb.LGBMRegressor(**params)
    m.fit(
        X.iloc[tr_idx], y.iloc[tr_idx],
        eval_set=[(X.iloc[vl_idx], y.iloc[vl_idx])],
        callbacks=[
            lgb.early_stopping(200, verbose=False),
            lgb.log_evaluation(500),
        ],
    )
    vp           = m.predict(X.iloc[vl_idx]).astype(np.float32)
    oof[vl_idx]  = vp
    preds       += m.predict(X_pred).astype(np.float32) / N_FOLD
    r            = rmse(y.iloc[vl_idx], vp)
    fold_scores.append(r)
    print(f"  Fold {fold}  RMSE={r:.5f}  "
          f"iters={m.best_iteration_}  t={time.time()-t0:.0f}s")
    del m
    gc.collect()

oof_score = rmse(y, oof)
print(f"\n  CV  mean: {np.mean(fold_scores):.5f}  "
      f"± {np.std(fold_scores):.5f}")
print(f"  OOF RMSE: {oof_score:.5f}")

# ── 7. PLOTS ──────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(16, 4))

# rebuild feature importances from last fold model — re-train tiny proxy
# (m was deleted above to save RAM; use fold-5 result still in scope)
# Note: if you want importances, remove 'del m' above
axes[0].set_title("Feature Importances\n(re-enable by keeping last fold model)")
axes[0].text(0.5, 0.5, "del m saved RAM\nsee code note",
             ha="center", va="center", transform=axes[0].transAxes)

step = max(1, len(y) // 50_000)
axes[1].scatter(y.values[::step], oof[::step],
                alpha=0.2, s=2, color="#B71C1C")
mn, mx = float(y.min()), float(y.max())
axes[1].plot([mn, mx], [mn, mx], "k--", lw=1)
axes[1].set_title(f"OOF Pred vs Actual   RMSE={oof_score:.4f}")
axes[1].set_xlabel("Actual TVT")
axes[1].set_ylabel("Predicted TVT")

axes[2].hist(y.values - oof, bins=80,
             color="#1B5E20", edgecolor="white", lw=0.3)
axes[2].axvline(0, color="black", lw=1.5)
axes[2].set_title("OOF Residuals")

plt.tight_layout()
plt.savefig("diagnostics.png", dpi=110)
plt.show()

# ─────────────────────────────────────────────────────────────────
# STEP 8 ─ SUBMISSION  (CORRECT ID ALIGNMENT)
#
#  v4 fell into Strategy-C (row-order) because WELL_NAME+ROW_IDX
#  weren't available.  Now they are, so we do:
#      id  =  WELL_NAME + "_" + ROW_IDX
#  then merge onto sample order.
#
#  Sample format example:
#      id              tvt
#      000d7d20_1442   NaN
# ─────────────────────────────────────────────────────────────────

print("\n[8] Building submission.csv …")

sample  = pd.read_csv(SAMPLE_SUB)
id_col  = sample.columns[0]
tvt_col = sample.columns[1]
print(f"    Sample columns : {sample.columns.tolist()}")
print(f"    Sample rows    : {len(sample):,}")
print(sample.head(3))

# ★ construct IDs that match sample_submission exactly
df_pred = df_pred.copy()
df_pred[id_col] = (
    df_pred["WELL_NAME"].astype(str)
    + "_"
    + df_pred["ROW_IDX"].astype(str)
)

sub_raw = pd.DataFrame({
    id_col:  df_pred[id_col].values,
    tvt_col: preds,
})

# align to sample order
sub = sample[[id_col]].merge(sub_raw, on=id_col, how="left")

n_miss = int(sub[tvt_col].isna().sum())
if n_miss > 0:
    fill = float(np.nanmean(preds))
    sub[tvt_col].fillna(fill, inplace=True)
    print(f"    ⚠  {n_miss} missing IDs — filled with mean {fill:.4f}")
    print("       Check that WELL_NAME / ROW_IDX match sample IDs!")
else:
    print(f"    ✓  All {len(sub):,} IDs matched — zero NaN")

# ── hard assertions ───────────────────────────────────────────────
assert len(sub) == len(sample),          \
    f"Row mismatch: {len(sub)} vs {len(sample)}"
assert sub[tvt_col].isna().sum() == 0,   \
    "NaN predictions remain"
assert list(sub.columns) == [id_col, tvt_col], \
    f"Wrong columns: {sub.columns.tolist()}"

sub.to_csv("submission.csv", index=False)
print("\n✅  submission.csv ready!")
print(sub.head(10).to_string())
print(f"\n  Stats  mean={preds.mean():.4f}  std={preds.std():.4f}  "
      f"min={preds.min():.4f}  max={preds.max():.4f}")
print(f"\n  OOF RMSE = {oof_score:.5f}")
print("""
╔══════════════════════════════════════════════════╗
║  BEFORE SUBMITTING                               ║
║  1. Settings → Accelerator → GPU T4 ✓           ║
║  2. Settings → Internet    → OFF ✓              ║
║  3. Save Version → Run All                       ║
║  4. submission.csv appears → Submit button on   ║
╚══════════════════════════════════════════════════╝
""")
```
